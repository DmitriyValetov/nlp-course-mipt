### Выполнение 2

```python
from nltk.metrics import jaccard_distance
from itertools import combinations
```

```python
def get_map(unique_symbols):
    res = pd.DataFrame(index=langs, columns=langs)
    for lang1, lang2 in combinations(langs, 2):
        jd = np.round(jaccard_distance(set(unique_symbols[lang1]), set(unique_symbols[lang2])), 2)
        res.loc[lang1][lang2] = jd
        res.loc[lang2][lang1] = jd
    for lang in langs:
        res.loc[lang][lang] = 1

    return res
```
